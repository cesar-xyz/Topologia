```python
# Import necessary libraries and modules
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from sklearn.metrics import pairwise_distances
```

```python
# Function to create a Vietoris-Rips complex given a distance matrix and radius
def create_simplicial_complex(D, r):
    """
    Input: distance matrix and not negative radius
    Output: networkx graph
    """

    Graph = nx.Graph()
    Graph.add_nodes_from(list(range(len(D))))
    edge_list = np.argwhere(D <= r)
    Graph.add_edges_from(edge_list)

    # Remove self-loops
    Graph.remove_edges_from(nx.selfloop_edges(Graph))

    return Graph, edge_list
```

```python
from numpy import sqrt

X = np.array([[1, 0], [-1, 0], [0, 1], [0, -1], [sqrt(2) / 2, sqrt(2) / 2], [-sqrt(2) / 2, sqrt(2) / 2],
              [-sqrt(2) / 2, -sqrt(2) / 2], [sqrt(2) / 2, -sqrt(2) / 2]])

for n in X:
    circle = plt.Circle(n, 0.5, color='red', fill=False)
    plt.gca().add_artist(circle)
    circle = plt.Circle(n, 1, color='green', fill=False)
    plt.gca().add_artist(circle)
    circle = plt.Circle(n, 2, color='blue', fill=False)
    plt.gca().add_artist(circle)

# Plot the combined points
plt.plot(X[:, 0], X[:, 1], 'o')
plt.axis('equal')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Calculate pairwise distance matrices for x and y
dist = pairwise_distances(X)
plt.imshow(dist)
plt.show()
```

```python
# Create and plot Vietoris-Rips complex for radius 0.5
r = 0.5
G, Edges = create_simplicial_complex(dist, r)
nx.draw_kamada_kawai(G)
for n in G.nodes():
    circle = plt.Circle(X[n], r, color='red', fill=False)
    plt.gca().add_artist(circle)
plt.xlim([-2, 2])
plt.ylim([-2, 2])

plt.show()
```
